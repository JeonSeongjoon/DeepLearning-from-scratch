```python
import numpy as np

D = 5
N = 7
```

```python
# Sum node
x = np.random.rand(1,D)
y = np.repeat(x, N, axis=0)

dy = np.random.rand(N, D)
dx = np.sum(dy, axis=0, keepdims=True)

print('dy:', dy,'\n')
print('dx:', dx)
print()
print("dy.shape : {}\ndx.shape : {}".format(dy.shape, dx.shape))
```

```python
# repeat node
x = np.random.rand(N, D)
y = np.sum(x, axis=0, keepdims=True)

dy = np.random.rand(1, D)
dx = np.repeat(dx, N, axis=0)

print('dy:', dy,'\n')
print('dx:', dx)
print()
print("dy.shape : {}\ndx.shape : {}".format(dy.shape, dx.shape))
```

```python
class MulLayer:
    def __init__(self):
        self.x = None
        self.y = None

    def forward(self, x, y):
        self.x = x          # save the input values for the backward propagation process
        self.y = y
        out = x * y

        return out

    def backward(self, dout):
        dx = dout * self.y
        dy = dout * self.x

        return dx, dy
    

apple_price = 100
apple_num = 2
tax = 1.1

mul_apple_layer = MulLayer()
mul_tax_layer = MulLayer()

o1 = mul_apple_layer.forward(apple_price, apple_num)
o2 = mul_tax_layer.forward(o1, tax)

print(o2)

dprice = 1
dapple_set, dtax = mul_tax_layer.backward(dprice)
dapple_price, dapple_num = mul_apple_layer.backward(dapple_set)

print(dapple_price, dtax)
```

```python

# I want to get the differential value of f(x) = x^2 at x = 1

def numerical_diff_forward(x):
    h = 1e-4
    f_x = lambda x: x**2
    return (f_x(x+h) - f_x(x)) / h

def numerical_diff_central(x):      #numerical differentiation
    h = 1e-4
    f_x = lambda x: x**2
    return (f_x(x+h) - f_x(x-h)) / 2*h

def analytic_diff(x):       #analytic differentiation
    return 2*x

print(f"numerical diffentiation: {numerical_diff_forward(x=1)}")
print(f"numerical diffentiation: {numerical_diff_central(x=1)}")
print(f"analytic diffentiation: {analytic_diff(x=1)}")
```

```python
import copy
import numpy as np

def gradient(f, x):
    '''
    The function that gets the gradient value of vector-valued funciton f at x

    f : A vector-valued function f
    x : A point x 
    '''
    tmp_x = copy.deepcopy(x)
    grad = np.zeros_like(x)
    h = 1e-4

    for i in range(x.size):
        hej = np.zeros_like(x) 
        hej[i] = h

        fx1 = f(tmp_x + hej) 
        fx2 = f(tmp_x - hej)

        grad[i] = (fx1-fx2)/(2*h)

    return grad

f = lambda x: x[0] + x[1]**2 + x[2]**3 + x[3]**4 + x[4]**5
x = np.array([3.0,1.0,5.0,6.0,2.0])

print(gradient(f, x))
```

```python
def numerical_gradient(f, x):
    '''
    The function that gets the gradient value of vector-valued funciton f at x
    
    f : A vector-valued function f
    x : A point x 
    '''

    tmp_x = copy.deepcopy(x)
    grad = np.zeros_like(x)
    h = 1e-4

    it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])
    while not it.finished:
        idx = it.multi_index
        hej = np.zeros_like(x) 
        hej[idx] = h

        fx1 = f(tmp_x + hej) 
        fx2 = f(tmp_x - hej)

        grad[idx] = (fx1-fx2)/(2*h)
        it.iternext()

    return grad

f = lambda x: x[0] + x[1]**2 + x[2]**3 + x[3]**4 + x[4]**5
x = np.array([3.0,1.0,5.0,6.0,2.0])

print(numerical_gradient(f, x))
```
